## ModelTraining

```python
# splitting data into X_train,y_train,X_test,y_test

X_train = train.drop('RUL',axis=1)
y_train = train['RUL']
X_test = test.drop('RUL', axis =1)
y_test = test['RUL']
```

```python
for i in [X_train,y_train,X_test,y_test]:
    print(i.shape)
```

```python
# Scaling

from sklearn.preprocessing import RobustScaler  # Uses IQR for scaling
scaler = RobustScaler()
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
#pip install xgboost
```

```python
# importing some necessary models

from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
# creating models list

models = {
 'Linear Regression': LinearRegression(),
 'Support Vector Machine': SVR(),
 'K Nearest Neighbors' : KNeighborsRegressor(),
 'Random Forest': RandomForestRegressor(),
 'Gradient Boosting': GradientBoostingRegressor(),
 'XGBoost': XGBRegressor()
 }
```

```python
# defining a function to calculate mae, rmse, r2 square 

def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
# checking for model performance

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train_scale,y_train)

    # Make Predictions
    y_pred=model.predict(X_test_scale)

    # validation(test) score
    mae, rmse, r2_square = evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    #model_list.append(list(models.keys())[i])

    # printing the results
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score:",r2_square)

    #r2_list.append(r2_square)
    
    print('\n')
```

```python
# applying threshold to check if performance increases

train['RUL'].describe()
```

* For predictive maintenance, the 25th percentile is often a useful threshold for early warnings!
* Catches engines in early degradation (balances false alarms and missed failures).
* here 25th percentile is 51


```python
train['RUL'][train['RUL'] > 51] = 51
test['RUL'][test['RUL'] > 51] = 51
```

```python
X_train_scale = scaler.fit_transform(X_train)
X_test_scale = scaler.transform(X_test)
```

```python
# checking for model performance after threshold limit
results = []
for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train_scale,y_train)

    # Make Predictions
    y_pred=model.predict(X_test_scale)

    # validation(test) score
    mae, rmse, r2_square = evaluate_model(y_test,y_pred)

    model_names = list(models.keys())[i]

    # printing the results
    results.append({
        'Model': model_names,
        'MAE': mae,
        'RMSE': rmse,
        'R2': r2_square,
    })

results = pd.DataFrame(results)
results
```

```python
# hyperparameter tuning using randomised search for svm

from sklearn.model_selection import RandomizedSearchCV
param_dist = {
    'C': [5],
    'epsilon': [0.1],
    'kernel': ['rbf'],
    'degree': [3],
    'gamma': ['auto'],
    'coef0': [0.0],
    'shrinking': [True],
    'tol': [0.001],
    'cache_size': [500]
 }

rscv = RandomizedSearchCV(SVR(), param_distributions=param_dist,n_iter=100, cv=5, n_jobs=-1, verbose=2, random_state=42)
rscv.fit(X_train_scale, y_train)
y_pred = rscv.predict(X_test_scale)
mae, rmse, r2_square = evaluate_model(y_test,y_pred)

print("RMSE:",rmse)
print("MAE:",mae)
print("R2 score:",r2_square)
```

```python
# hyperparameter tunning using randomised search for random forest

rfr = RandomForestRegressor()

param_dist = {
    'n_estimators': [200],      # Number of trees
    'max_depth': [20],          # Max depth of trees
    'min_samples_split': [10],   # Min samples to split a node
    'min_samples_leaf': [5],    # Min samples per leaf
    'max_features': ['sqrt'],   # Features to consider per split
    'bootstrap': [True]         # Bootstrap sampling
}

rscv = RandomizedSearchCV(rfr, param_distributions=param_dist,n_iter=100, cv=5, n_jobs=-1, verbose=2, random_state=42)
rscv.fit(X_train_scale, y_train)
y_pred = rscv.predict(X_test_scale)
mae, rmse, r2_square = evaluate_model(y_test,y_pred)

print("RMSE:",rmse)
print("MAE:",mae)
print("R2 score:",r2_square)
```

* random forest performs slightly better
